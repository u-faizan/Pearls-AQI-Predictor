```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/raw_data_islamabad_20241224-20251224.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Missing values
df.isnull().sum()
```

```python
# Convert the column to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])
# Check date range
print(f"Start: {df['timestamp'].min()}")
print(f"End: {df['timestamp'].max()}")
print(f"Total days: {(df['timestamp'].max() - df['timestamp'].min()).days}")
```

```python
# Check for duplicates
df.duplicated().sum()
```

#**Ploting PM2.5 over time**

```python
# Plot PM2.5 over time
plt.figure(figsize=(12, 4))
plt.plot(df['timestamp'], df['pm2_5'])
plt.title('PM2.5 Over Time')
plt.xlabel('Date')
plt.ylabel('PM2.5 (g/m)')
plt.show()
```

# **Distribution of pollutants**

```python
# Distribution of pollutants
df[['pm10', 'pm2_5', 'ozone', 'nitrogen_dioxide']].hist(bins=50, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

# **Correlation heatmap**

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
corr = df[['pm10', 'pm2_5', 'carbon_monoxide', 'nitrogen_dioxide',
           'sulphur_dioxide', 'ozone', 'temperature_2m',
           'relative_humidity_2m', 'wind_speed_10m']].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
