```python
import pandas as pd
import numpy as np
```

```python
# Load data
df = pd.read_csv('../data/processed/processed_aqi.csv')
df.head()
```

```python
# Select numeric columns only
df_numeric = df.select_dtypes(include=[np.number])
```

```python
df_numeric.corr()
```

```python
# Compute correlations with AQI
corr_with_aqi = (
    df_numeric
    .corr()['aqi']
    .sort_values(ascending=False)
)
```

```python
# Display results
print("Top 15 features positively correlated with AQI:\n")
display(corr_with_aqi.head(15).to_frame(name='correlation'))
```

```python
print("\nBottom 10 features negatively correlated with AQI:\n")
display(corr_with_aqi.tail(10).to_frame(name='correlation'))
```
