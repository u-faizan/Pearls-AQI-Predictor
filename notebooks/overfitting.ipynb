```python
summary = []

for model_name, model_metrics in metrics.items():
    train_r2 = model_metrics['train']['r2']
    val_r2 = model_metrics['validation']['r2']
    test_r2 = model_metrics['test']['r2']
    gap = train_r2 - test_r2
    
    summary.append({
        'Model': model_name,
        'Train R²': round(train_r2, 4),
        'Val R²': round(val_r2, 4),
        'Test R²': round(test_r2, 4),
        'Train–Test Gap': round(gap, 4)
    })

df_summary = pd.DataFrame(summary)
print("\n OVERFITTING SUMMARY")
display(df_summary)
```
